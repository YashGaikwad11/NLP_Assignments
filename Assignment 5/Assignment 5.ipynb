Import WordNet

```python
from nltk.corpus import wordnet as wn
```

Synonymy (same meaning)

```python
word = "happy"
synsets = wn.synsets(word)

synonyms = set()
for syn in synsets:
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms of", word, ":", synonyms)
```

Hypernymy (Finding Hypernyms)

```python
word = "dog"
syn = wn.synsets(word)[0]
hypernyms = syn.hypernyms()

for h in hypernyms:
    print("Hypernym:", h.name())
```

Hyponymy (Finding Hyponyms)

```python
word = "vehicle"
syn = wn.synsets(word)[0]
hyponyms = syn.hyponyms()

for h in hyponyms[:5]:
    print("Hyponym:", h.name())
```

Apply to Real Text

```python
text = "The boy is happy"
words = text.split()

for w in words:
    synsets = wn.synsets(w)
    if synsets:
        print("\nWord:", w)
        print("Synonyms:", [l.name() for l in synsets[0].lemmas()])
        print("Hypernyms:", [h.name() for h in synsets[0].hypernyms()])
```
