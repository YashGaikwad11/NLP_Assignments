```python
text = "Artificial Intelligence is transforming healthcare and education in India."
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

sentences = sent_tokenize(text)
words = word_tokenize(text)

print("Sentences:", sentences)
print("Words:", words)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
filtered_words = [w for w in words if w.lower() not in stop_words and w.isalpha()]

print("Filtered Words:", filtered_words)
```

```python
from nltk import pos_tag

pos_tags = pos_tag(filtered_words)
print("POS Tags:", pos_tags)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered_words]

print("Lemmatized Words:", lemmatized)
```

```python
from nltk import ne_chunk

ner = ne_chunk(pos_tags)
print("Named Entities:")
print(ner)
```

```python
from nltk.probability import FreqDist

freq = FreqDist(filtered_words)
print("Most Common Words:", freq.most_common(5))
```
